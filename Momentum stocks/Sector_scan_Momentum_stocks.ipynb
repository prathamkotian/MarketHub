```python
# Further filter the DataFrame to include only the stocks name in the stock list

stockList = stockDF.iloc[:,0].tolist()[1:]
stockList
```

```python
# Convert the 'expiry' column to datetime format and extract the date part
symboldf['expiry'] = pd.to_datetime(symboldf['expiry']).apply(lambda x: x.date())

# Filter the DataFrame to include only rows
symboldf = symboldf[(symboldf.exchange == 'NSE_EQ') & (symboldf.last_price.notnull()) & (symboldf.lot_size == 1) &  (symboldf.tick_size == 0.05) ]

# include only rows where the 'tradingsymbol' is in the 'stockList'
symboldf = symboldf[symboldf.tradingsymbol.isin(stockList)]
symboldf.reset_index(inplace=True,drop=True)
symboldf
```

```python
def getHistData(instrument):
    try:
        instrument = urllib.parse.quote(instrument)
        today = datetime.now()
        # Format today's date and the date 100 days ago
        to_date =datetime.strftime(today,'%Y-%m-%d')  
        from_date = datetime.strftime(today- timedelta(days=100),'%Y-%m-%d') 
        
        # Construct the URL for the API request
        url = f'https://api.upstox.com/v2/historical-candle/{instrument}/day/{to_date}/{from_date}' 
        
        # Make the API request and get the response as JSON
        candleRes = requests.get(url, headers={'accept': 'application/json'}).json()   
        
        # Convert the JSON response to a Pandas DataFrame
        candleData = pd.DataFrame(candleRes['data']['candles'])
        
        # Rename the columns of the DataFrame
        candleData.columns = ['date','open','high','low', 'close','vol','oi']
        candleData =  candleData[['date','open','high','low', 'close']]
        candleData['date'] = pd.to_datetime(candleData['date']).dt.tz_convert('Asia/Kolkata')
        candleData['date'] = candleData.apply(lambda x: x.date.date(), axis=1)
        candleData.sort_values(by = 'date', inplace=True)
        return candleData
    except Exception as e:
        print(f"Exception when calling HistoryApi->get_intra_day_candle_data {e}")
```

```python
# Function to calculate relative strength

def calculate_relative_strength(base_symbol_close, comparative_symbol_close, length):
    
    # Calculate the relative price changes for the base symbol over the specified length
    base_relative = base_symbol_close / base_symbol_close.shift(length)
    comparative_relative = comparative_symbol_close / comparative_symbol_close.shift(length)

    # Calculate the relative strength between the base and comparative symbols
    relative_strength = base_relative / comparative_relative -1 
    
    # Round the relative strength to 2 decimal places
    relative_strength_rounded = relative_strength.round(2)
    return relative_strength_rounded
```

```python
# Get historical data for Nifty 500 and Nifty Pharma indices
indexData = getHistData('NSE_INDEX|Nifty 500')
sectorData = getHistData('NSE_INDEX|Nifty Pharma')
#sectorData = getHistData('NSE_INDEX|Nifty PSU Bank')
filterList = []
for i in  symboldf.index:
    try:
        symInfo = symboldf.loc[i]
        symbolName =symInfo['tradingsymbol']
        
        stockCandleData = getHistData(symInfo['instrument_key'])
        # Merge stock data with index data
        finalData = stockCandleData.merge(indexData[['date','close']],on = 'date',how='left',suffixes=("", "_index"))
        # Merge the result with sector data
        finalData = finalData.merge(sectorData[['date','close']],on = 'date',how='left',suffixes=("", "_sector"))
        
        # Calculate relative strength compared to index and sector
        finalData['RS_INDEX'] = calculate_relative_strength(finalData['close'], finalData['close_index'], length=50)
        finalData['RS_SECTOR'] = calculate_relative_strength(finalData['close'], finalData['close_sector'], length=50)
        
        # Get the most recent data
        recentData = finalData.iloc[-1]
        recentData.loc['sym'] = symbolName
        
        # Append the recent data to the filter list
        filterList.append(recentData)
        
    except Exception as e :
        print(f"Error in {symboldf.loc[i].to_dict()}")
        traceback.print_exc() 
```

```python
# Convert the filter list to a DataFrame

sectorRSdf = pd.DataFrame(filterList)
sectorRSdf
```

```python
# Calculate the rank by summing RS_INDEX and RS_SECTOR

sectorRSdf['rank'] = sectorRSdf['RS_INDEX'] + sectorRSdf['RS_SECTOR']
sectorRSdf
```

### The Top most stock is with most momentum in that perticular sector

```python
# Sort the DataFrame by rank in descending order

sectorRSdf.sort_values(by = 'rank',ascending=False)
```
